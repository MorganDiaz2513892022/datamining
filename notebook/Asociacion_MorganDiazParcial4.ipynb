```python
#importacion de librerias
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#link de correlacion
url = "https://raw.githubusercontent.com/MorganDiaz2513892022/MorganDiaz-Parcial4/refs/heads/main/data/clave_A_asociacion.csv"

df = pd.read_csv(url)

df.head()
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
#verificacion de valores nulos, duplicados y tipos de datos
print(df.info())
print(df.isnull().sum())
print("Duplicados:", df.duplicated().sum())
```

```python
#preparacion de datos
df_bool = df.astype(bool)

print(df_bool.head())
```

```python
#aplicacion de algoritmo apriori
frequent_itemsets = apriori(
    df_bool,
    min_support=0.10,
    use_colnames=True
)

frequent_itemsets = frequent_itemsets.sort_values(
    by="support",
    ascending=False
)

frequent_itemsets.head(10)
```

```python
#generacion de reglas de asociacion
#Si un cliente compra A, también suele comprar B
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.5
)

rules = rules.sort_values(
    by=["confidence", "lift"],
    ascending=False
)

rules.head(10)
```

```python
#mostrar las reglas mas relevantes
#se muestran las reglas con mayor confianza y lift
top10 = rules[[
    'antecedents',
    'consequents',
    'support',
    'confidence',
    'lift'
]].head(10)

print(top10)
```
